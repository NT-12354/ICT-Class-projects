```python
weight = float(input("please enter your weight(in kg): "))
height = float(input("please enter your height(in cm): "))

BMI = float(weight/height * height)
rounded_BMI = round(BMI, 2)
print(rounded_BMI)


if BMI < 18.5: 
    print("underweight")
elif 18.5 <= BMI <= 24.9:
    print("Normal weight")
elif 25 <= BMI <= 29.9:
    print("Overweight")
elif BMI >= 30:
    print("obese")
else: 
    print("please enter a valid value")
```
