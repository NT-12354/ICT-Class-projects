```python

def atm_withdrawal(): 
    account_balance = 1000
    print("account_balance = 1000")
    withdrawal = float(input("Please enter your withdrawal amount: "))
    if withdrawal > account_balance:
        print ("Insufficient funds")
    elif withdrawal == account_balance:
        print("Balance will be 0")
        print(account_balance == 0)
    else:
        print ("withdrawal successful")
        print (account_balance - withdrawal)


atm_withdrawal()
```
