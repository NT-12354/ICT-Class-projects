```python
income = float(input("please enter your income: "))
def tax_calculator():
    
    if income <= 10000:
        tax = 0
    elif income <= 20000:
        tax = (income - 10000) * 0.1
    elif income <= 50000:
        tax = (10000 * 0) + (10000 * 0.1) + (income - 20000) * 0.2
    else:
        tax = (10000 * 0) + (10000 * 0.1) + (30000 * 0.2) + (income - 50000) * 0.3

    net_income = income - tax

    print("Total tax owed:", tax)
    print("Net income after tax:", net_income)

tax_calculator()

def low():
    print("0%")
def llow():
    print("15%")
def mlow():
    print("20%")
def hlow():
    print("25%")
def middle():
    print("30%")
def high():
    print("35%")

if income <= 2000:
    low()
elif income > 2001 and income <= 4000:
    llow()
elif income > 4001 and income <= 7000:
    mlow()
elif income > 7001 and income <= 10000:
    hlow()
elif income > 10001 and income <= 14000:
    middle()
elif income > 14000:
    high()

    
```
